```python
import pandas as pd

energy_use = pd.read_csv('../input/hourly-energy-consumption/AEP_hourly.csv', parse_dates=['Datetime'], index_col=['Datetime'])

# Use functions from https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DatetimeIndex.html
# to get date features

energy_use['year'] = energy_use.index.year
energy_use['month'] = energy_use.index.month
energy_use['day_of_month'] = energy_use.index.day
energy_use['day_of_week'] = energy_use.index.dayofweek

energy_use.head()
```

```python
from matplotlib import pyplot as plt
energy_use.groupby('year').mean().AEP_MW.plot()
plt.show()
```

```python
from matplotlib import pyplot as plt
energy_use.groupby('month').mean().AEP_MW.plot()
plt.show()
```
